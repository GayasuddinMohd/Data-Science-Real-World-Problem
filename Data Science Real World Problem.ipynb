# Pivoting the data

```python
pivot_table=pd.pivot_table(data, values='Quantity', index='CustomerID', columns='InvoiceDate')
```

```python
pivot_table
```

# Predicting the Quantity

Example of predicting the Quantity if, the Date and customer id is present in the pivot tabele(Dataset)

```python
customer_id=17850.0
```

```python
customer_id
```

```python
desired_date=pd.to_datetime('2010-12-01 08:26:00')
```

```python
desired_date
```

```python
# Retrieve the quantity for the provided customer ID and date
```

```python
pivot_table.loc[customer_id,desired_date]
```

```python
if customer_id in pivot_table.index and desired_date in pivot_table.columns:
    quantity_prediction = pivot_table.loc[customer_id, desired_date]
    print(f"Predicted quantity for Customer ID {customer_id} on {desired_date}: {quantity_prediction}")
else:
    print(f"No data found for Customer ID {customer_id} on {desired_date}.")
```
